## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Number of Booking in each Hotels

````python
# Chart - 1
plt.figure(figsize=(12,6))
sns.countplot(data=hotel_new, x='Hotel')
plt.xlabel('Hotel')
plt.ylabel('No of Booking')
plt.title('Total Number of booking in Hotels', fontsize=20)
plt.show()
````

Observation:



*   City Hotel has more numbers of booking compare to Resort Hotel.  


*  Resort Hotel has 32000(approx) booking and City Hotel has 50000(approx) booking.



#### Country-wise Analysis

````python
country_data= hotel_new['Country'].value_counts().head(10)
country_data
````

````python
# Chart - 2 visualization code
plt.figure(figsize=(15,4))
sns.barplot(data=hotel_new, x=country_data.index, y=country_data.values)
plt.xlabel('Country')
plt.ylabel('No of Guest')
plt.title('Country Wise Analysis', fontsize=20)
plt.show()
````

In this visualisation I took top 10 country from where more number of guests visited to both the hotels.
Most of the customers come from Portugal, Great Britain, France and Spain.

#### Booking percentage of Hotel

````python
hotel_grpby= pd.DataFrame((hotel_new.groupby('Hotel').size()/hotel_new.shape[0])*100).reset_index().rename(columns={0:'Booking Percentage'})

print(hotel_grpby)
````

````python
# Chart - 3 visualization code
plt.figure(figsize=(10,5))
sns.barplot(x=hotel_grpby['Hotel'], y=hotel_grpby['Booking Percentage'])
plt.xlabel('Hotels')
plt.ylabel('Booking Percentage')
plt.title('Hotel Wise Analysis of Booking', fontsize=20)
plt.show()
````

Around 60% bookings are for City hotel and 40% bookings are for Resort hotel.

#### Stay at the Hotel

````python
not_canceled = hotel_new[hotel_new['Canceled'] == 0]
s1 = not_canceled[not_canceled['total_night_stay'] < 15]
````

````python
# Chart - 4 visualization code
plt.figure(figsize = (15,6))
sns.countplot(x = s1['total_night_stay'], hue = s1['Hotel'])
plt.xlabel('Total Number of Nights')
plt.ylabel('Number of Guests')
plt.title('Number of stay in Hotels', fontsize=20)
plt.show()
````

Most common stay length is less than 4 days and generally people prefer City hotel for short stay, but for long stays, Resort Hotel is preferred.

Hotel should work on to increase total stay in Resort hotel to increase revenue.

#### Weekdays & Weekend Analysis

````python
hotel_new
````

````python
# Chart - 5 visualization code
plt.figure(figsize = (20,8))

plt.subplot(1,2,1)
sns.boxplot(x = "Market Segment", y = "Stay in Week Night", data = hotel_new, hue = "Hotel", palette = 'Set1');
plt.xlabel('Market Segment')
plt.ylabel('Stay in Weekdays Night')
plt.title('Guest stay in Weekdays Nights')

plt.subplot(1,2,2)
sns.boxplot(x = "Market Segment", y = "Stay in Weekend Night", data = hotel_new, hue = "Hotel", palette = 'Set1');
plt.xlabel('Market Segment')
plt.ylabel('Stay in Weekends Night')
plt.title('Guest stay in WeekEnds Night')

plt.show()
````

````python
plt.figure(figsize = (15,8))
sns.boxplot(x = "Market Segment", y = "total_night_stay", data = hotel_new, hue = "Hotel", palette = 'Set1')
plt.xlabel('Market Segment')
plt.ylabel('Stay in Weekdays & Weekends')
plt.title('Guest stay in WeekEnds & Weekdays Nights', fontsize=20)

plt.show()
````

It can be seen that most of the groups are normally distributed, some of them have high skewness. Looking at the distribution, it can be concluded that most people do not seem to prefer to stay at the hotel for more than 1 week. But it seems normal to stay in resort hotels for up to 12–13 days. Although this changes according to the segments, staying longer than 15 days certainly creates outliers for each segment. If the total time feature was created by summing up the weekend and week nights, this would be clearer, but it can be clearly seen when looking at the two visualizations together.

#### Months wise Analysis

````python
# Chart - 6 visualization code
plt.figure(figsize=(15,8))

sns.countplot(x=hotel_new['Month'].sort_values(ascending=True), data=hotel_new, hue='Hotel')
plt.xlabel('Stay in Months')
plt.xticks(rotation=50)
plt.ylabel('Number of Guest')
plt.title('Months wise Analysis', fontsize=20)
plt.show()
````

Above insight shows that in the month of August and July, there is more number of stay in compare to others. There is also negative insight, but hotel can use this insight to arrange everything in advance and welcome their guest in the best way possible and hotel can also run some promotional offer in these 2 months to attract more guests. In the winter months like Jan, Feb, Nov, Dec, each hotels should decrease the price of their rooms and add some comfortable(Hot Gyesers, Room Heater, etc) to attract customers.
